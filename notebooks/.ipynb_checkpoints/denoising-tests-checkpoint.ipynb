```python
import numpy as np
from scipy import stats
```

```python
data = np.loadtxt("../data/Langevin_1D.txt")
```

```python
data
```

```python
data.shape
```

```python
samples_per_bin = 100
bin_size = int(len(data) / samples_per_bin)
bins = np.array_split(data, bin_size)
bins[0]
```

```python
stats.ks_2samp(bins[0], bins[1])
```

```python
ks_data = [stats.ks_2samp(bins[i], bins[i + 1]) for i in range(len(bins) - 1)]
ks_stats = [x.statistic for x in ks_data]
ks_pvals = [x.pvalue for x in ks_data]
ks_data[0]
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
fig.set_size_inches(20,10)
# ax.plot(np.arange(0, samples_per_bin * len(ks_data),samples_per_bin), ks_stats, c='r')
# ax.plot(np.arange(0, samples_per_bin * len(ks_data),samples_per_bin), ks_pvals, c='b')
ax.plot(np.arange(len(data)), data, c='black')
for index, pval in enumerate(ks_pvals):
    if pval < 0.05:
        ax.vlines(samples_per_bin * index, -1.5, 1.5, colors='red')
```

```python
# now we do the smoothing
# f(t) = avg(k-nearest(window(t)))
# so, subset the data around t according to window size
# find the k nearest neighbors for the data point at t
# average those neighbors and call that average your smooth point at t

def knearest(val, series, k=2):
    distances = enumerate([abs(val - x) for x in series])
    sorted_dists = sorted(distances, key=lambda x:x[1])
    indices = [dist[0] for dist in sorted_dists][0:k]
    values = [series[index] for index in indices]
    return np.average(values)

knearest(bins[0][1], bins[0], k = 5)
half_window = 50


smooth_data = [knearest(data[t], data[(t - half_window):(t + half_window)], k=50) for t in range(len(data))]
plt.clf()
fig, ax = plt.subplots()
fig.set_size_inches(20,7.5)
ax.plot(np.arange(len(data)), data, c='r')
ax.plot(np.arange(len(smooth_data)), smooth_data, c='b')
```
