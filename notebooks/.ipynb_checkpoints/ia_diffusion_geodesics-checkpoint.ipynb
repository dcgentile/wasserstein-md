```python
import total
import os
import numpy as np
import seaborn as sns
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import geopandas
from libpysal import weights
from gerrychain import Graph

img_dir = os.path.join(".", "geodesic_stills")
```

```python
filename = "./IA_counties/IA_counties.shp"
ia_graph = Graph.from_file(filename)
ia_counties = geopandas.read_file(filename)
ia_adj=nx.adjacency_matrix(ia_graph).toarray()
centroids = np.column_stack((ia_counties.centroid.x, ia_counties.centroid.y))
queen = weights.Queen.from_dataframe(ia_counties)
graph = queen.to_networkx()
positions = dict(zip(graph.nodes, centroids))
num_nodes = len(graph.nodes)
```

```python
def compute_graph_metric(graph):
    graph_dists = dict(nx.shortest_path_length(graph))
    sorted_dists = {k:sorted(v.items()) for k, v in graph_dists.items()}
    num_nodes = len(graph.nodes)
    dist_mat = np.zeros((num_nodes, num_nodes))
    for i in range(num_nodes):
        for j in range(num_nodes):
            dist_mat[i,j] = sorted_dists[i][j][1]
    
    return dist_mat

def iter_experiment(its_lo, its_hi, its_step, measures, cost, epsilon, steps):
    for its in range(its_lo, its_hi, its_step):
        tot_geodesic = total.build_geodesic(measures, cost, epsilon, loss="sqeuc", method="total", steps=steps, iters=its)
        pot_geodesic = total.build_geodesic(measures, cost, epsilon, method="", steps=steps)
        #tot_ratio = np.divide(tot_geodesic, pot_geodesic, out=np.zeros_like(pot_geodesic))
        tot_ratio = tot_geodesic/pot_geodesic
        tot_rel_err = abs(np.ones((num_nodes, steps)) - tot_ratio)
        fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
        fig.set_size_inches(16,8)
        sns.heatmap(tot_geodesic, ax=ax1)
        sns.heatmap(pot_geodesic, ax=ax2)
        sns.heatmap(tot_rel_err, ax = ax3)
        return fig, tot_rel_err
```

```python
def comparison_video(measures, cost, epsilon, steps, its, filename, metric):
    tot_geodesic = total.build_geodesic(measures, cost, epsilon, method="total", steps=steps, iters=its)
    pot_geodesic = total.build_geodesic(measures, cost, epsilon, method="", steps=steps)
    fig, (ax1, ax2) = plt.subplots(1, 2)
     
    dir_name = "comparison-anims/"
    still_dir = os.path.join(img_dir, dir_name)
    if not os.path.isdir(still_dir):
        os.mkdir(still_dir)
     
    def update(frame):
        nx.draw(graph, positions, ax=ax1, node_size=50, node_color=tot_geodesic[:, frame])
        nx.draw(graph, positions, ax=ax2, node_size=50, node_color=pot_geodesic[:, frame])
        ax1.set_title(
            f"TOTAL: {metric}\n Timestep : {frame + 1}"
            )
        ax2.set_title(
            f"POT: {metric}\n Timestep : {frame + 1}"
            )
    ani = animation.FuncAnimation(fig=fig, func=update, frames=steps, interval=100)
    ani.save(os.path.join(dir_name, filename), writer='ffmpeg')
```

```python
#c0, c1 = 34, 9
c0, c1 = np.random.randint(num_nodes), np.random.randint(num_nodes)
graph_metric_cost = compute_graph_metric(graph)
cost = graph_metric_cost
mu_0 = total.random_geographic_concentration(cost, center=c0)
mu_1 = total.random_geographic_concentration(cost, center=c1)
measures = np.stack((mu_0, mu_1)).T
epsilon = 0.1
steps = 50
```

```python
cost = graph_metric_cost ** 2
its = 256
for _ in range(10):
    c0, c1 = np.random.randint(num_nodes), np.random.randint(num_nodes)
    filename = f"ia-squared-graph-metric-geodesic-{c0}-to-{c1}.mp4"
    comparison_video(measures, cost, epsilon, steps, its, filename, "Graph Metric")
```
